## Disease Background: 

*  Pulmonary fibrosis is a chronic lung disease in which lung tissue becomes thickened, stiff, and scarred. In idiopathic pulmonary fibrosis, or IPF, the exact cause is unknown, but the result is progressive damage to the lung interstitium and impaired gas exchange. As fibrosis worsens, the lungs become less compliant, making breathing more difficult and reducing oxygen delivery to the blood.

* Standard of care treatments: 
As of currently there is no cure/treatment for pulmonary fibrosis, most current "treatments" are used to alleviate the symtoms and improve qaulity of life, Current current includes: 
- Antifibrotic medications
Antifibrotic medication are the most commonly used because they target the disease process by slowing scar formation. while the medication does not treat the fibrosis it self it does slow down the tissue scarring. 
- Oxygen therapy
One of the most common symtoms of lung fibrosis is rapid shallow breathing, nausea, and syncope, ehich is the cause of lock of oxygen entering the lungs for circulation. Oxygen therapy works by increasing the amount of oxygen the body receives and thus improving the qaulity of life. 

Used when oxygen levels are low. It does not treat fibrosis but is very effective at relieving symptoms and improving quality of life. 
Lung transplant:

* Biological mechanisms (anatomy, organ physiology, cell & molecular physiology):

Pulmonary fibrosis happens when the alveoli, normally thin and contain a high concentration of elastin for gas exchange, become thickened and scarred from excess collagen buildup. This makes the lungs stiff and harder to expand, so oxygen cannot move into the bloodstream as easily. As a result, breathing takes more effort and less oxygen reaches the body, leading to symptoms like shortness of breath and fatigue.
